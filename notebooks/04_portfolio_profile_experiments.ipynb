```python
import pandas as pd

from ai_models.feature_builder import build_feature_table
from ai_models.quality_score_model import run_quality_score_model

feature_result = build_feature_table(
    fundamentals_path="data/fundamentals_cache.parquet",
    prices_path="data/prices_cache.parquet",
    treasury_path="data/treasury_yields_cache.parquet",
    benchmark_ticker="SPY",
)

features = feature_result.features.reset_index()
quality = run_quality_score_model(features)
universe = features.merge(quality, on="Ticker", how="left")
universe.head()
```

```python
conservative = universe.sort_values(["QualityScore", "Volatility_63D"], ascending=[False, True]).head(15)
balanced = universe.sort_values(["QualityScore", "Momentum_12M"], ascending=[False, False]).head(15)
aggressive = universe.sort_values(["Momentum_12M", "QualityScore"], ascending=[False, False]).head(15)

conservative[["Ticker", "QualityScore", "Volatility_63D", "Momentum_12M"]].head(10)
```
