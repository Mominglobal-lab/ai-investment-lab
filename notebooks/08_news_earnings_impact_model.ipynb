# News and Earnings Impact Model Prototype

Goal: design the structure for company event inputs and experiment with simple impact scoring before productionizing a news pipeline.

```python
import pandas as pd
```

```python
# Replace this synthetic sample with real fetched earnings/news later.
events = pd.DataFrame(
    [
        {"Ticker": "NVDA", "EventType": "Earnings", "Headline": "Revenue beats expectations", "Sentiment": 0.9, "Magnitude": 0.8},
        {"Ticker": "AAPL", "EventType": "Guidance", "Headline": "Soft near-term guidance", "Sentiment": -0.5, "Magnitude": 0.6},
        {"Ticker": "MA", "EventType": "News", "Headline": "Stable cross-border volume growth", "Sentiment": 0.4, "Magnitude": 0.3},
    ]
)
events
```

```python
events["ImpactScore"] = events["Sentiment"] * events["Magnitude"]
events = events.sort_values("ImpactScore", ascending=False)
events
```

```python
portfolio = pd.DataFrame(
    [
        {"Ticker": "NVDA", "Weight": 0.25},
        {"Ticker": "AAPL", "Weight": 0.25},
        {"Ticker": "MA", "Weight": 0.10},
    ]
)
portfolio_impact = portfolio.merge(events, on="Ticker", how="left")
portfolio_impact["WeightedImpact"] = portfolio_impact["Weight"] * portfolio_impact["ImpactScore"].fillna(0.0)
portfolio_impact
```
